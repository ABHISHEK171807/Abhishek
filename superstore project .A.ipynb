### HYPOTHESIS


```python
##.H1. TECHOLOGY PRODUCTS HAVE THE HIGHEST PROFIT MARGIN COMPARED
TO OTHER PRODUCT CATEGORIES
```

```python
data.columns
```

```python
cat_profit = data.groupby ('Category')['Profit'].sum()
cat_profit
```

```python
cat_profit = data.groupby ('Category')['Profit'].sum()
cat_profit. plot (kind='bar')
plt.title("profit by category")
plt.xlabel("Category")
plt.ylabel("Total Profit")
plt.show()

##CONCLUSION : THE HYPOTHESIS IS SUPPORTED AS TECHNOLOGY PRODUCTS HAVE THE 
# THE HIGHEST PROFIT MARGIN OF THE THREE CATEGORIES
```

### H2. THE EAST REGION HAS THE HIGHEST SALES COMPARED TO OTHER REGIONS


```python
data.columns
```

```python
reg_sales=data.groupby('Region')['Sales'].sum()
reg_sales
```

```python
reg_sales=data.groupby('Region')['Sales'].sum()
reg_sales.plot(kind="bar")
plt.title("Total sales by region")
plt.xlabel("Region")
plt.ylabel("'Sales")
plt.show()

## conclusion : the hypothesis is not supported as the central region has 
## the highest sales
```

### H3. SALES ARE HIGHER DURING CERTAIN MONTHS OF THE YEAR


```python
data.columns
```

```python
pd.DatetimeIndex(data['Order Date']).month

```

```python
data['Order Month']=pd.DatetimeIndex(data['Order Date']).month
```

```python
month_sales=data.groupby('Order Month')['Sales'].sum()
month_sales.plot(kind='line')
plt.title("Total Sales by Month")
plt.xlabel("Month")
plt.ylabel("Total Sales")
plt.show()
```

#### H4. ORDEERS WITH SAME DAY SHIPPING HAVE THE LOWEST RATE OF RETURNED PRODUCTS.


```python
data.columns
```

```python
data.groupby('Ship Mode').size()
```

```python
total_orders_by_shipping_mode=data.groupby('Ship Mode').size()

return_orders_by_shipping_mode=data[data['Profit']<0].groupby('Ship Mode').size()

returned_per_by_shipping_mode=(return_orders_by_shipping_mode/total_orders_by_shipping_mode)*100
print(returned_per_by_shipping_mode)

returned_per_by_shipping_mode.plot(kind="bar")
plt.title("Return per by Shipping Mode")
plt.xlabel("Shipping Mode")
plt.ylabel("Return per")
plt.show()

## conclusion : THE HYPOTHESIS IS SUPPORTED AS ORDERS WITH SAME DAY
## SHIPPING HAVE THE LOWEST RATE OF RETURNED PRODUCT
```
